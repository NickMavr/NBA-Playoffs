```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read the csv and turn it into a panda dataframe
df = pd.read_csv('NBA1.csv')
# show the 10 first rows to see all the features 
df.head(10)
```

```python
# describe the dataset by count,mean,std,min,25%,50%,75%,max
df.describe()
```

```python
# give the info about the dataset
df.info()
```

```python
# create target column which has the target 
df['target'] = df['Playoffs']
# count how many True or False defects we have to see the balance of the set 
df['target'].value_counts()  
```

```python
# remove the irrelevant features from the dataset 
A = df.drop(['Playoffs','target','Rk','Team','MP','PF','FGA','FTA','Age','3PA','2PA','PW','PL','W','L'], axis=1)
X = np.array(A)

# convert all to float and make them an array
X = X.astype(np.float32)
Y = np.array(df['target'])

# see the shape of the arrays to check if the arrays has the right dimensions
print(X.shape)
print(Y.shape)
```

```python
# import all the models,scorers,cross_validate,pipeline and scalers
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from imblearn.metrics import geometric_mean_score
from sklearn.metrics import make_scorer

from sklearn.pipeline import Pipeline

from sklearn.model_selection import cross_validate, ShuffleSplit
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC
from sklearn.model_selection import StratifiedKFold


# Define classifiers
classifiers = {
    'Logistic Regression': LogisticRegression(solver = 'lbfgs', max_iter = 5000, random_state = 1,penalty = 'l2', C = 1.5 , class_weight = 'balanced'),
    'Decision Tree': DecisionTreeClassifier(criterion = 'gini', max_depth = 4, random_state = 1, min_samples_split = 20),
    'SVC Linear': LinearSVC(dual = False, random_state = 1, penalty = 'l1', max_iter = 5000, tol = 1e-5, fit_intercept = True)
}

# Define scalers
scalers = {
    'No Normalization': None,
    'Standardization': StandardScaler(),
    'Min-max normalization': MinMaxScaler()
}

# define scores
scorers = {
'accuracy_score': make_scorer(accuracy_score),
'f1_score': make_scorer(f1_score, average = 'weighted'),
'precision_score': make_scorer(precision_score, average = 'weighted', zero_division = 0),
'recall_score': make_scorer(recall_score, average = 'weighted')}

plt.figure(figsize=(15,3), dpi = 150)
bar_width = 0.25

# names of the classifiers for the labes of the bars
clf_names = ['Logistic Regression', 'Decision Tree', 'SVC Linear']
x_label = np.arange(3) 

fig, ax = plt.subplots(4, 1,figsize = (10, 14))
plt.subplots_adjust(hspace = 0.6)

# random permutations cross-validation(Shuffle & Split) for better results 
cv = ShuffleSplit(n_splits = 5, test_size = 0.2, random_state = 1)

# Define cross-validation strategy (Stratified K-Fold)
# skf = StratifiedKFold(n_splits = 5, shuffle = True, random_state = 1)

i = 0
# Loop through classifiers and scalers
for clf_name, clf in classifiers.items():
    j = 0
    for scaler_name, scaler in scalers.items():
            
        # define pipeline
        pipe = Pipeline([(scaler_name, scaler), (clf_name,clf)])
        
        # cross validate with the help of the pipeline
        cv_results = cross_validate(pipe, X, Y, cv = cv, scoring = scorers, return_train_score = True, error_score = 'raise')

        # define scorers
        accuracy = cv_results['test_accuracy_score'].mean()
        f1 = cv_results['test_f1_score'].mean()
        precision = cv_results['test_precision_score'].mean()
        recall = cv_results['test_recall_score'].mean()
        
        print(clf_name,scaler_name,
              'precision:',precision,
              'recall:',recall,
              'accuracy:',accuracy,
              'f1:',f1,)  
        
        if scaler_name == 'Standardization':
            color = '#7E2F7E'
        elif scaler_name == 'Min-max normalization':
            color = '#77AC70'
        else:
            color = '#0077BD'
        if i == 0:
            ax[0].bar(i + j * bar_width, accuracy, width = bar_width, color = color, label = scaler_name)
            ax[0].text(i + j * bar_width, accuracy, f"{round(accuracy, 3)}", ha = 'center', va = 'bottom', fontsize = 6)
            ax[1].bar(i + j * bar_width, f1, width = bar_width, color = color, label = scaler_name)
            ax[1].text(i + j * bar_width, f1, f"{round(f1, 3)}", ha = 'center', va = 'bottom', fontsize = 6)
            ax[2].bar(i + j * bar_width, recall, width = bar_width, color = color, label = scaler_name)
            ax[2].text(i + j * bar_width, recall, f"{round(recall, 3)}", ha = 'center', va = 'bottom', fontsize = 6)
            ax[3].bar(i + j * bar_width, precision, width = bar_width, color = color, label = scaler_name)
            ax[3].text(i + j * bar_width, precision, f"{round(precision, 3)}", ha = 'center', va = 'bottom', fontsize = 6)
        else:
            ax[0].bar(i + j * bar_width, accuracy, width = bar_width, color = color)
            ax[0].text(i + j * bar_width, accuracy, f"{round(accuracy, 3)}", ha = 'center', va = 'bottom', fontsize = 6)
            ax[1].bar(i + j * bar_width, f1, width = bar_width, color = color)
            ax[1].text(i + j * bar_width, f1, f"{round(f1, 3)}", ha = 'center', va = 'bottom', fontsize = 6)
            ax[2].bar(i + j * bar_width, recall, width = bar_width, color = color)
            ax[2].text(i + j * bar_width, recall, f"{round(recall, 3)}", ha = 'center', va = 'bottom', fontsize = 6)
            ax[3].bar(i + j * bar_width, precision, width = bar_width, color = color)
            ax[3].text(i + j * bar_width, precision, f"{round(precision, 3)}", ha = 'center', va = 'bottom', fontsize = 6)
        j += 1
    i += 1

ax[0].set_ylabel('Accuracy')
ax[0].set_xticks(x_label + bar_width, clf_names)
ax[0].set_ylim(0, 1)
ax[0].legend(loc = 'upper left', bbox_to_anchor = (0.5, 1.15), fontsize = 'xx-small', ncol = 3)

ax[1].set_ylabel('F1')
ax[1].set_xticks(x_label + bar_width, clf_names)
ax[1].set_ylim(0, 1)
ax[1].legend(loc = 'upper left', bbox_to_anchor = (0.5, 1.15), fontsize = 'xx-small', ncol = 3)

ax[2].set_ylabel('Recall')
ax[2].set_xticks(x_label + bar_width, clf_names)
ax[2].set_ylim(0, 1)
ax[2].legend(loc = 'upper left', bbox_to_anchor = (0.5, 1.15), fontsize = 'xx-small', ncol = 3)

ax[3].set_ylabel('Precision')
ax[3].set_xticks(x_label + bar_width, clf_names)
ax[3].set_ylim(0, 1)
ax[3].legend(loc = 'upper left', bbox_to_anchor = (0.5, 1.15), fontsize = 'xx-small', ncol = 3)

plt.show()
```
